```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm
import matplotlib
import scipy

matplotlib.font_manager._rebuild()
```

```python
import matplotlib.font_manager as fm
from matplotlib import rcParams
rcParams['font.family'] = 'Times New Roman'
rcParams['font.sans-serif'] = ['Times New Roman']
fontsize=19
font = fm.FontProperties(family='Times New Roman', weight=None, style='normal', size=fontsize)
```

```python
def plot_cls_empirical_dist_density(x, source, target, filename=None):
    df = pd.DataFrame.from_dict(
        {
        'x': x+x,
        'sample size': source+target,
        'domain': ['source']*len(source) + ['target']*len(target)
        })
    figsize = (3.3, 4.5)
    fig, ax = plt.subplots(num=None, figsize=figsize, dpi=70, facecolor='w', edgecolor='k')
    
    # sns.set_context("talk")
    paper_rc = {'lines.linewidth': 4.2, 'lines.markersize': 12, "axes.labelsize":22, "axes.titlesize":8,}                  
    sns.set_context("talk", rc = paper_rc)             
    sns.set_palette("Set2")
    #sns.set_style("ticks", {'font.family':'serif', 'font.serif':'Times New Roman'})

    sns.lineplot(x='x', y='sample size', hue='domain', data=df)
    
    
    
    ax.set(xticklabels=[])
    #ax.set(yticklabels=[])
    ax.tick_params(axis=u'both', which=u'both',length=0)

    #plt.xlim(0, 64)
    #plt.ylim(48, 70)

    plt.legend( ('Source', 'Target'), loc='best', prop=font)
    plt.xlabel('Class labels', fontsize=fontsize)
    plt.ylabel('Probability density function', fontsize=fontsize)
    plt.tight_layout()
    if filename is not None:
        plt.savefig(f'../plots/{filename}.pdf', bbox_inches = 'tight', pad_inches = 0.05)
```

```python
alpha=100
fig, ax = plt.subplots(1, 1)

num_of_class = 65
x = np.linspace(1, 5, num_of_class)
target = scipy.stats.genpareto.pdf(x, alpha).tolist()
source = np.flip(scipy.stats.genpareto.pdf(x, alpha)).tolist()

#plot_cls_empirical_dist_density(x.tolist(), source, target, filename='paredo_dist')
```

```python
num_of_class = 10
alpha=1
x = np.linspace(1, 5, num_of_class)
target = scipy.stats.genpareto.pdf(x, alpha).tolist()
target = [t/sum(target) for t in target]
source = np.flip(np.array(target)).tolist()
plot_cls_empirical_dist_density(x.tolist(), source, target, filename=None)
```
