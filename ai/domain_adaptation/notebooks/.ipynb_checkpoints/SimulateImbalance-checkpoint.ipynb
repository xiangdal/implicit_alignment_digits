```python
import collections

def load_data(path):
    group_ids = collections.defaultdict(list)
    subclass_ids = collections.defaultdict(list)
    
    with open(path, 'r') as f:
        data = f.readlines()
        
    for i, line in enumerate(data):
        line = line.split(' ')
        group_ids[line[1].strip()].append(i)
        subclass_ids[line[2].strip()].append(i)
        
    return group_ids, subclass_ids
```

```python
group_ids, subclass_ids = load_data(
    '/Users/xjiang/GoogleDrive/Colab_Experiments/meta-domain-adaptation/ai/domain_adaptation/data/DomainNet_ParentClasses/clipart_test.txt')
```

```python
def show_stats(group_ids, subclass_ids):
    print('group stats:')
    for g in sorted(group_ids, key=lambda x: int(x)):
        print(f'\t{g}:\t{len(group_ids[g])}')
    print('subclass stats:')
    for sub in sorted(subclass_ids, key=lambda x: int(x)):
        print(f'\t{sub}:\t{len(subclass_ids[sub])}')
```

```python
show_stats(group_ids, subclass_ids)
```
