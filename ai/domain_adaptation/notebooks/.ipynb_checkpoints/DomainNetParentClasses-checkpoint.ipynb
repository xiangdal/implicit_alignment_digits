```python
def retrieve_names(s):
    s = s.split('\n')
    s = [si.split('&')[0].strip().replace(' ', '_') for si in s if si]
    return s
    
```

```python
class_structure = OrderedDict()
for i, parent_class in enumerate(raw_data):
    class_structure[i] = retrieve_names(raw_data[parent_class])
    if parent_class == 'Office':  # remove bandage and nail from the office group
        class_structure[i].remove('bandage')
        class_structure[i].remove('nail')
```

```python
print(f'total number of sub classes: {sum([len(v) for v in class_structure.values()])}')
```

```python
classname_to_id_dict = {}
for k, v in class_structure.items():
    for subclass_name in v:
        classname_to_id_dict[subclass_name] = k
classname_to_id_dict
```

```python
from os import listdir
from os.path import isfile, join
path = '/Users/xjiang/GoogleDrive/Colab_Experiments/meta-domain-adaptation/ai/domain_adaptation/data/DomainNet/'
files = [f for f in listdir(path) if isfile(join(path, f)) and f.endswith('.txt')]
```

```python
cnt = 0
for k,v in classname_to_id_dict.items():
    if v <= 1:
        cnt += 1
cnt
```

```python
def get_subtype_indices(parent_labels):
    cnt = 0
    classname_to_subtype_dict = {}
    for k,v in classname_to_id_dict.items():
        if v in parent_labels:
            classname_to_subtype_dict[k] = cnt
            cnt += 1
    return classname_to_subtype_dict

def process_one_img(s, parent_labels=None):
    s = s.split(' ')
    filepath, label = s[0], s[1]
    class_name = filepath.split('/')[2]
    parent_label = classname_to_id_dict[class_name]
    
    if parent_labels is None:
        return f'{filepath} {parent_label} {label}\n'
    elif parent_label in parent_labels:
        classname_to_subtype_dict = get_subtype_indices(parent_labels)
        return f'{filepath} {parent_label} {classname_to_subtype_dict[class_name]}\n'
    else:
        return None

def mergeClasses(input_filename, output_filename, parent_labels=None):
    with open(input_filename, 'r') as f_read:
        content = f_read.readlines()
        new_content = [process_one_img(c, parent_labels) for c in content]
        
        new_content = [c for c in new_content if c]
        
        with open(output_filename, 'w') as f_write:
            f_write.writelines(new_content)

# process all files
from os import listdir
from os.path import isfile, join
path = '/Users/xjiang/GoogleDrive/Colab_Experiments/meta-domain-adaptation/ai/domain_adaptation/data/DomainNet/'
files = [f for f in listdir(path) if isfile(join(path, f)) and f.endswith('.txt')]
for f in files:
    mergeClasses(
        join('/Users/xjiang/GoogleDrive/Colab_Experiments/meta-domain-adaptation/ai/domain_adaptation/data/DomainNet/', f),
        join('/Users/xjiang/GoogleDrive/Colab_Experiments/meta-domain-adaptation/ai/domain_adaptation/data/DomainNet_ParentClasses_Subset/', f),
        parent_labels=[0, 1]
    )
```
