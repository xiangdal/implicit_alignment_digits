```python
def analyze_predictions(
    pkl_path='../pickles/domainnet.sourceonly.pkl',
    domain_file='../data/DomainNet_ParentClasses/clipart_test.txt',
    pdf_name=f'../plots/subtype_var_parent_sourceonly.pdf',
    title='source only (no adaptation)'
):
    if not pkl_path.endswith('.pkl'):
        pkl_path += '.pkl'
    
    x = np.load(pkl_path, allow_pickle=True)
    pred = np.argmax(x, axis=1)
    domain_info = image_index.parse_domain_file(domain_file)
    d = {
        'parent_labels': domain_info.label_list,
        'sub_labels': domain_info.subtype_label_list,
        'accuracy': pred == domain_info.label_list
    }
    df = pd.DataFrame(data=d)
    parent_stats = df.groupby('parent_labels').mean()
    stats = df.groupby('sub_labels').mean()
    
#     ax = sns.boxplot(x="parent_labels", y="accuracy", data=stats)
#     ax = sns.swarmplot(x="parent_labels", y="accuracy", data=stats, color=".25")
#     plt.xlabel('parent classes')
#     plt.ylabel('subtype accuracy distribution', fontsize=fontsize)
#     plt.title(title,
#               fontdict={'fontsize': rcParams['axes.titlesize']})
#     plt.tight_layout()
#     plt.savefig(pdf_name, bbox_inches = 'tight', pad_inches = 0.01)
#     import pdb; pdb.set_trace()
    std = stats.groupby('parent_labels').std()
    print(f'accuracy:\n\t{df.accuracy.mean()*100:.2f}')
    print(f'avg accuracy by group:\n\t{parent_stats.accuracy.mean()*100:.2f}')
    print(f'average subtype accuracy:\n\t{stats.accuracy.mean()*100:.2f}')  #.quantile()
    print(f'average within-group variance:\n\t{std.mean()[0]*100:.2f}')  #.quantile()
    
    
    return stats
```

# hypothesis test

```python
def ttest_groups(a, b):
    std_a = a.groupby('parent_labels').std().accuracy.to_numpy()
    std_b = b.groupby('parent_labels').std().accuracy.to_numpy()
    return scipy.stats.ttest_rel(std_a, std_b)
```

```python
mi_stats.groupby('parent_labels').std().accuracy.to_numpy()
```

```python
mi_implicit_flat_stats.groupby('parent_labels').std().accuracy.to_numpy()
```

```python
ttest_groups(mi_implicit_flat_stats, mi_stats)
```

```python
ttest_groups(mi_implicit_hierarchical_stats, mi_stats)
```
